```python
import tensorflow as tf
print(tf.__version__)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# buat data dan persamaan
N = 1000
X = np.random.random((N, 2)) * 6 - 3 #Distribusi antara (-3, +3)
Y = np. cos(2*X[:,0]) + np.cos(3*X[:,1])
```

### Implementasi dari fungsi
y = cos(2x1) + cos(3x2)

```python
#visualisasi
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
```

```python
#Membuat model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, input_shape=(2,), activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
# Mengompile dan training
opt = tf.keras.optimizers.Adam(0.01)
model.compile(optimizer=opt, loss='mse')
train = model.fit(X, Y, epochs=100)
```

```python
#Visualisasi training loss
plt.plot(train.history['loss'], label='loss')
```

```python
# Visualisasi prediksi
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(X[:,0], X[:,1], Y)

# surface plot
line = np.linspace(-3, 3, 50)
xx, yy = np.meshgrid(line,line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
Yhat = model.predict(Xgrid).flatten()
ax2.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Yhat, linewidth=0.2, antialiased=True)
plt.show()
```

```python
#Ekstraploasi
fig3 = plt.figure()
ax3 = fig3.add_subplot(111, projection='3d')
ax3.scatter(X[:,0], X[:,1], Y)

line = np.linspace(-5, 5, 50)
xx, yy = np.meshgrid(line,line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
Yhat = model.predict(Xgrid).flatten()
ax3.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Yhat, linewidth=0.2, antialiased=True)
plt.show()
```
